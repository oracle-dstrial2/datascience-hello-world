# Linear Regression Predictor

This exploration loads a publically available data set: the Boston Housing Data.  It then splits the data into test and training data, uses a linear regression model to generate predictions, and plots the results.

```python
# import machine learning and plotting packages
from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt

# initialize the linear regression model
lr = linear_model.LinearRegression()

# load the data
boston = datasets.load_boston()
# set the test data
y = boston.target

# generates predictions by cross validation:
predicted = cross_val_predict(lr, boston.data, y, cv=10)

# plot the results
fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

The chart above shows that a linear regression model provides a reasonable fit for this data set.
